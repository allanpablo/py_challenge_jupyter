# Visualização de Dados com Pandas e Matplotlib

Nesta aula, vamos aprender a criar:

- **Gráfico de dispersão** → relação entre variáveis
- **Gráfico de boxplot** → distribuição, mediana e outliers
- **Gráfico de linhas** → evolução temporal

Esses gráficos facilitam a análise e comunicação de padrões nos dados.


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Gerar dados simulados
np.random.seed(42)
df = pd.DataFrame({
    "Idade": np.random.randint(18, 60, 50),
    "Altura": np.random.normal(1.75, 0.08, 50),
    "Peso": np.random.normal(70, 10, 50),
    "Data": pd.date_range("2023-01-01", periods=50, freq="W")
})

df["IMC"] = df["Peso"] / (df["Altura"] ** 2)
df.head()
```

## Gráfico de Dispersão

Utilizado para mostrar a **relação entre duas variáveis**, ideal para identificar padrões ou outliers.


```python
plt.figure(figsize=(8,5))
plt.scatter(df["Idade"], df["IMC"], color='darkorange', alpha=0.7)
plt.title("Gráfico de Dispersão - Idade x IMC")
plt.xlabel("Idade")
plt.ylabel("IMC")
plt.grid(True)
plt.show()
```

## Gráfico de Boxplot

Mostra a **distribuição dos dados**, incluindo:

- Mediana
- Quartis (Q1, Q3)
- Outliers


```python
plt.figure(figsize=(6,5))
plt.boxplot(df["IMC"], vert=True)
plt.title("Boxplot do IMC")
plt.ylabel("IMC")
plt.grid(True)
plt.show()
```

## Gráfico de Linhas

Muito usado para representar **dados ao longo do tempo**, como séries temporais, tendências e evolução.


```python
plt.figure(figsize=(10,5))
plt.plot(df["Data"], df["Peso"], marker='o', linestyle='-', color='royalblue')
plt.title("Gráfico de Linhas - Peso ao Longo do Tempo")
plt.xlabel("Data")
plt.ylabel("Peso (kg)")
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Conclusão

- O **gráfico de dispersão** é excelente para visualizar relações entre variáveis.
- O **boxplot** ajuda a entender a distribuição e detectar valores extremos.
- O **gráfico de linhas** mostra claramente a evolução de uma variável no tempo.

Essas representações facilitam a análise exploratória de dados e a comunicação visual dos resultados.


# Histograma com Pandas e Matplotlib

Nesta aula, vamos:

- Garantir que o Matplotlib esteja instalado
- Usar o método `hist()` do Pandas para criar histogramas
- Gerar um gráfico da distribuição da variável `Idade`
- Salvar o gráfico em imagem
- Visualizar em tela cheia no VS Code


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Simular dados de idade
np.random.seed(0)
df = pd.DataFrame({
    "Idade": np.random.randint(18, 70, size=100)
})

df.head()
```

## Histograma da Idade

O histograma mostra a **distribuição da frequência** de uma variável.  
O eixo X mostra as faixas de valores (bins), e o eixo Y mostra quantas vezes cada faixa aparece.


```python
df["Idade"].hist(bins=10, color='skyblue', edgecolor='black')
plt.title("Histograma da Idade")
plt.xlabel("Idade")
plt.ylabel("Frequência")
plt.grid(True)
plt.show()
```

# Gráficos de Barras com Pandas

Nesta aula, vamos criar:

- Gráficos de barras **verticais** (`plot.bar`)
- Gráficos de barras **horizontais** (`plot.barh`)
- A partir de dados organizados com Pandas
- Incluindo ordenação, rótulos e formatação


```python
import pandas as pd
import matplotlib.pyplot as plt

# Dicionário com dados mais complexos
vendas = {
    "Categoria": ["Eletrônicos", "Roupas", "Alimentos", "Móveis", "Livros", "Beleza", "Brinquedos"],
    "Valor (R$)": [85000, 42000, 56000, 30000, 18000, 25000, 15000],
    "Quantidade": [120, 300, 500, 80, 150, 210, 95]
}

df = pd.DataFrame(vendas)
df
```

```python
# Ordenar por valor (decrescente)
df_valor = df.sort_values("Valor (R$)", ascending=False)

plt.figure(figsize=(10,6))
plt.bar(df_valor["Categoria"], df_valor["Valor (R$)"], color="mediumseagreen")
plt.title("Valor Total Vendido por Categoria")
plt.xlabel("Categoria")
plt.ylabel("Valor em R$")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Ordenar por quantidade (crescente)
df_qtd = df.sort_values("Quantidade", ascending=True)

plt.figure(figsize=(10,6))
plt.barh(df_qtd["Categoria"], df_qtd["Quantidade"], color="cornflowerblue")
plt.title("Quantidade de Itens Vendidos por Categoria")
plt.xlabel("Quantidade")
plt.ylabel("Categoria")
plt.tight_layout()
plt.show()
```

```python
# Adicionar rótulos nos gráficos verticais
plt.figure(figsize=(10,6))
bars = plt.bar(df_valor["Categoria"], df_valor["Valor (R$)"], color="darkorange")
plt.title("Valor Vendido por Categoria (com rótulos)")
plt.xlabel("Categoria")
plt.ylabel("Valor (R$)")
plt.xticks(rotation=45)

# Inserir rótulos
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1000, f"{yval:,.0f}", ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

## Conclusão

- O método `plot.bar()` é ideal para **comparar valores entre categorias**.
- Podemos usar `sort_values()` para **ordenar os dados** antes de plotar.
- O método `plot.barh()` é útil quando as categorias têm nomes longos ou o número de barras é grande.
- Gráficos de barras são ótimos para **visualização de comparações diretas**.


# Gráfico de Dispersão com Pandas e Matplotlib

Nesta aula, vamos:

- Criar um gráfico de dispersão entre **idade** e **altura**
- Ver se existe **correlação positiva**
- Traçar uma **reta de ajuste (regressão linear simples)**
- Discutir como esse gráfico é útil para análise e previsão


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import pearsonr

# Gerar dados simulados
np.random.seed(42)
idades = np.random.randint(18, 60, 50)
alturas = 1.55 + (idades * 0.005) + np.random.normal(0, 0.03, 50)

df = pd.DataFrame({
    "Idade": idades,
    "Altura": alturas
})

df.head()
```

## Gráfico de Dispersão

Usamos a função `plt.scatter()` para mostrar a relação entre duas variáveis.  
Aqui vamos relacionar `Idade` (eixo X) com `Altura` (eixo Y).


```python
plt.figure(figsize=(8,5))
plt.scatter(df["Idade"], df["Altura"], color="mediumblue", alpha=0.7)
plt.title("Dispersão: Idade x Altura")
plt.xlabel("Idade (anos)")
plt.ylabel("Altura (m)")
plt.grid(True)
plt.show()
```

## Correlação e Linha de Tendência

Vamos calcular o **coeficiente de correlação de Pearson** e traçar uma **reta de regressão linear**:

\[
\text{altura} = a \cdot \text{idade} + b
\]

A reta nos ajuda a prever a altura com base na idade.


```python
# Correlação de Pearson
corr, _ = pearsonr(df["Idade"], df["Altura"])
print(f"Coeficiente de Correlação de Pearson: {corr:.4f}")

# Regressão linear (reta de melhor ajuste)
x = df["Idade"]
y = df["Altura"]
a, b = np.polyfit(x, y, deg=1)  # y = a*x + b

# Plot
plt.figure(figsize=(8,5))
plt.scatter(x, y, label="Dados", color="dodgerblue", alpha=0.6)
plt.plot(x, a*x + b, color="red", label=f"y = {a:.3f}x + {b:.3f}")
plt.title("Dispersão com Reta de Ajuste")
plt.xlabel("Idade")
plt.ylabel("Altura")
plt.legend()
plt.grid(True)
plt.show()
```

## Conclusão

- O **gráfico de dispersão** mostra a relação visual entre duas variáveis.
- A **correlação positiva** indica que, em geral, quanto maior a idade, maior a altura.
- A **reta de regressão** pode ser usada para prever valores (ex: altura estimada para uma idade).
- Esse tipo de gráfico é fundamental em análises exploratórias e em Machine Learning supervisionado.


# Boxplot com Pandas

O gráfico de caixa (boxplot) é uma representação visual da distribuição dos dados.  
Ele exibe:

- **Mediana** (linha no centro da caixa)
- **Quartis** (Q1 e Q3)
- **Limites inferior e superior** (whiskers)
- **Outliers** (pontos fora da caixa)

Nesta aula, vamos analisar as variáveis `Idade` e `Altura`.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Simular dados
np.random.seed(42)
idades = np.random.randint(18, 65, size=100)
alturas = np.random.normal(1.70, 0.08, size=100)

# Inserir alguns outliers nas alturas
alturas[5] = 2.10
alturas[15] = 1.40

df = pd.DataFrame({
    "Idade": idades,
    "Altura": alturas
})

df.head()
```

## Boxplot das variáveis

Vamos gerar um gráfico de caixa para `Idade` e `Altura`  
O método `df.boxplot(column=...)` permite visualizar uma ou mais variáveis.


```python
plt.figure(figsize=(8,6))
df.boxplot(column=["Idade", "Altura"])
plt.title("Boxplot - Idade e Altura")
plt.ylabel("Valores")
plt.grid(True)
plt.show()
```

## Interpretação

- A **mediana** aparece como uma linha dentro da caixa.
- A caixa vai do **1º quartil (Q1)** ao **3º quartil (Q3)**.
- Os "bigodes" representam o intervalo típico dos dados.
- Pontos fora dos bigodes são **outliers**, como uma altura de 2.10 m ou 1.40 m.


```python
# Visualizar altura isoladamente com mais detalhes
plt.figure(figsize=(6,5))
df.boxplot(column="Altura", grid=False)
plt.title("Boxplot - Altura")
plt.ylabel("Altura (m)")
plt.show()
```

## Conclusão

- O boxplot é uma ferramenta eficiente para **validar a distribuição dos dados**.
- Permite **comparar variáveis**, detectar **outliers** e entender **dispersão e assimetria**.
- Pode ser complementado com histogramas e medidas estatísticas como média, desvio padrão e curtose.


# Gráfico de Linhas com Dados Temporais

Nesta aula, vamos criar um **gráfico de linhas** com dados de **faturamento ao longo do tempo**.

Essa visualização nos ajuda a:

- Analisar **tendências**
- Identificar **picos de crescimento ou queda**
- Avaliar o comportamento de uma **série temporal**


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Simular datas e faturamento
datas = pd.date_range(start="2020-01-01", periods=36, freq="M")  # 3 anos mensais
faturamento = np.random.normal(loc=50000, scale=7000, size=36)
faturamento[[5, 12, 24]] += 20000  # picos
faturamento[[8, 20]] -= 15000      # vales

# Criar o DataFrame
df = pd.DataFrame({
    "Data": datas,
    "Faturamento": faturamento
})

# Garantir que 'Data' esteja no tipo datetime
df["Data"] = pd.to_datetime(df["Data"])
df.head()
```

## Gráfico de Linhas

Usamos `plot.line()` ou `df.plot()` com `kind="line"` para visualizar **a evolução do faturamento ao longo do tempo**.


```python
plt.figure(figsize=(12,6))
plt.plot(df["Data"], df["Faturamento"], marker='o', linestyle='-', color='teal')
plt.title("Faturamento Mensal da Empresa (2020–2022)")
plt.xlabel("Data")
plt.ylabel("Faturamento (R$)")
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Conclusão

- O gráfico de linhas é ideal para **visualizar séries temporais**.
- Podemos observar tendências de **crescimento, queda ou sazonalidade**.
- Com `pandas + matplotlib`, a criação é rápida e personalizável.
- Esse tipo de gráfico é útil para **análise de negócios e previsão com IA**.
