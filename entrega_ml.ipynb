```python
# Importar bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

# 1. Dados de entrada
dados_vendas = {
    'mes': ['Janeiro', 'Fevereiro', 'Mar√ßo', 'Abril', 'Maio', 'Junho', 'Julho', 
            'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro'],
    'vendas': [2000, 2200, 2300, 2500, 2600, 2700, 2800, 2900, 3000, 3100, 3200, 3300]
}

# 2. Transformar em DataFrame
df = pd.DataFrame(dados_vendas)

# 3. Criar uma coluna num√©rica para os meses
df['mes_num'] = df.index + 1  # Janeiro = 1, Fevereiro = 2, ..., Dezembro = 12

# 4. Separar vari√°veis para o modelo
X = df[['mes_num']]     # vari√°vel independente
y = df['vendas']        # vari√°vel dependente (target)

# 5. Treinar modelo de regress√£o linear
modelo = LinearRegression()
modelo.fit(X, y)

# 6. Prever vendas para Dezembro (m√™s 12)
dezembro_df = pd.DataFrame({'mes_num': [12]})
venda_prevista = modelo.predict(dezembro_df)[0]

print(f"Previs√£o de vendas para Dezembro: R$ {venda_prevista:.2f}")

# 7. Histograma das vendas
plt.figure(figsize=(8, 4))
plt.hist(df['vendas'], bins=5, edgecolor='black')
plt.title('Histograma das Vendas')
plt.xlabel('Valor da Venda')
plt.ylabel('Frequ√™ncia')
plt.grid(True)
plt.show()

# 8. Gr√°fico de Dispers√£o (scatter plot)
plt.figure(figsize=(8, 4))
plt.scatter(df['mes_num'], df['vendas'], color='blue', label='Dados reais')
plt.plot(df['mes_num'], modelo.predict(X), color='red', linestyle='--', label='Regress√£o Linear')
plt.title('Vendas por M√™s')
plt.xlabel('M√™s')
plt.ylabel('Vendas')
plt.xticks(df['mes_num'], df['mes'], rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
