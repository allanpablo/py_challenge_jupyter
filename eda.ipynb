# üìä Exploratory Data Analysis (EDA) com Pandas

Neste m√≥dulo, vamos explorar o conceito de **An√°lise Explorat√≥ria de Dados (Exploratory Data Analysis - EDA)** e como aplic√°-la em projetos reais de ci√™ncia de dados utilizando a biblioteca **Pandas**.

## üîç O que √© EDA?

EDA √© um **processo sistem√°tico** que nos permite:
- Compreender a estrutura e o conte√∫do de um conjunto de dados
- Identificar padr√µes, tend√™ncias, inconsist√™ncias e outliers
- Formular hip√≥teses para an√°lises mais profundas

## üõ†Ô∏è Etapas principais do EDA:

1. **Coleta e prepara√ß√£o dos dados**  
   - Capturar dados de diferentes fontes (CSV, banco de dados, APIs, etc.)  
   - Realizar limpeza, transforma√ß√£o e padroniza√ß√£o  
   - Tratar valores ausentes e dados at√≠picos

2. **Formula√ß√£o de hip√≥teses e an√°lise estat√≠stica**  
   - An√°lise **univariada**: comportamento de uma √∫nica vari√°vel  
   - An√°lise **bivariada**: rela√ß√£o entre duas vari√°veis (ex: correla√ß√£o)  
   - An√°lise **multivariada**: padr√µes entre m√∫ltiplas vari√°veis  
   - An√°lise **temporal**: evolu√ß√£o ao longo do tempo

3. **Visualiza√ß√£o e comunica√ß√£o de resultados**  
   - Usar gr√°ficos para facilitar a interpreta√ß√£o dos dados  
   - Comunicar insights de forma clara, adaptando-se ao p√∫blico-alvo  
   - Evitar jarg√µes t√©cnicos quando a audi√™ncia n√£o for t√©cnica

## üêº A Biblioteca Pandas

O **Pandas** √© uma biblioteca essencial para an√°lise de dados em Python. Ele fornece:
- Estruturas eficientes como **DataFrame** e **Series**
- Compatibilidade com arquivos CSV, Excel, SQL, JSON e mais
- Recursos para:
  - Filtragem de dados
  - Limpeza e transforma√ß√£o
  - C√°lculos agregados e estat√≠sticas
  - Organiza√ß√£o e an√°lise por grupos (groupby)
  - Integra√ß√£o com bibliotecas de visualiza√ß√£o como Matplotlib e Seaborn

## ‚úÖ Conclus√£o

A EDA √© uma etapa **fundamental para todo projeto de dados**.  
Com Pandas, conseguimos transformar dados crus em **informa√ß√µes valiosas** para tomada de decis√£o, modelagem e comunica√ß√£o com stakeholders.

# ‚ùì Aula: Lidando com Dados Ausentes em DataFrames

Nesta aula, vamos entender como identificar e tratar dados ausentes (missing values), que podem prejudicar a an√°lise ou o desempenho de modelos de intelig√™ncia artificial.

## üìå Tipos de dados ausentes

1. **MCAR (Missing Completely at Random)**: Aus√™ncias ocorrem de forma totalmente aleat√≥ria.
   - Ex: erro de coleta em sensores.
2. **MAR (Missing at Random)**: A aus√™ncia depende de outras vari√°veis observadas.
   - Ex: pessoas com menor renda tendem a n√£o responder perguntas sobre patrim√¥nio.
3. **NMAR (Not Missing at Random)**: A aus√™ncia depende do pr√≥prio valor ausente.
   - Ex: clientes que cancelaram o servi√ßo n√£o informam o motivo.

---


```python
# ‚úÖ 1. Detectar dados ausentes
# Verificar a quantidade de nulos por coluna
print(df_merged.isnull().sum())

df_merged.isna().sum()

# Verificar porcentagem de nulos
print(df_merged.isnull().mean().round(3) * 100)
```

```python
# ‚úÖ 2. Remover linhas ou colunas com dados ausentes
# Remover linhas com qualquer valor ausente
df_sem_nulos = df_merged.dropna()

# Remover colunas que possuem muitos nulos (exemplo: mais de 50%)
limite = len(df_merged) * 0.5
df_filtrado = df_merged.dropna(axis=1, thresh=limite)
```

```python
# ‚úÖ 3. Preencher (imputar) valores ausentes
# Preencher nulos com zero (ex: valores financeiros)
df_merged['totalcharges'] = df_merged['totalcharges'].fillna(0)

# Preencher com m√©dia da coluna
#df_merged['mensalidade'] = df_merged['mensalidade'].fillna(df_merged['mensalidade'].mean())
df_merged['monthlycharges'] = df_merged['monthlycharges'].fillna(df_merged['monthlycharges'].mean())

# Preencher com valor mais frequente (moda)
#df_merged['forma_pagamento'] = df_merged['forma_pagamento'].fillna(df_merged['forma_pagamento'].mode()[0])
df_merged['paymentmethod'] = df_merged['paymentmethod'].fillna(df_merged['paymentmethod'].mode()[0])
```

```python
# ‚úÖ 4. Verificar novamente
# Verificar se ainda restam valores ausentes
df_merged.isnull().sum()
```

## Identificar outliers com base no desvio interquartil (IQR) e aplicar estrat√©gias de tratamento.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Exemplo com a vari√°vel 'totalcharges'
coluna = 'totalcharges'

# 1. Calcular quartis e IQR
Q1 = df_merged[coluna].quantile(0.25)
Q3 = df_merged[coluna].quantile(0.75)
IQR = Q3 - Q1

# 2. Definir limites inferior e superior
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

# 3. Identificar outliers
outliers = df_merged[(df_merged[coluna] < limite_inferior) | (df_merged[coluna] > limite_superior)]

print(f"Quantidade de outliers detectados: {len(outliers)}")

# 4. Visualiza√ß√£o com boxplot
plt.figure(figsize=(8, 4))
plt.boxplot(df_merged[coluna].dropna())
plt.title(f'Distribui√ß√£o de {coluna}')
plt.ylabel(coluna)
plt.grid(True)
plt.show()
```

1. Remover os outliers


```python
df_sem_outliers = df_merged[(df_merged[coluna] >= limite_inferior) & (df_merged[coluna] <= limite_superior)]
```

2. Substituir por limite m√°ximo/m√≠nimo


```python
df_tratado = df_merged.copy()
df_tratado.loc[df_tratado[coluna] > limite_superior, coluna] = limite_superior
df_tratado.loc[df_tratado[coluna] < limite_inferior, coluna] = limite_inferior
```

 Agrupar os outliers √© uma alternativa √∫til quando voc√™ n√£o quer descart√°-los nem alterar seus valores, mas deseja analis√°-los separadamente ou marcar sua presen√ßa para modelagem.

Aqui est√° como voc√™ pode identific√°-los e marc√°-los como uma nova categoria, criando uma coluna is_outlier:

‚úÖ C√≥digo para marcar/agrupar outliers


```python
import pandas as pd

# Definindo a vari√°vel a ser analisada
coluna = 'totalcharges'

# 1. Calcular quartis e IQR
Q1 = df_merged[coluna].quantile(0.25)
Q3 = df_merged[coluna].quantile(0.75)
IQR = Q3 - Q1

# 2. Limites inferior e superior
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

# 3. Criar coluna bin√°ria indicando outlier (True/False)
df_merged['is_outlier_totalcharges'] = ~df_merged[coluna].between(limite_inferior, limite_superior)

# Verificar distribui√ß√£o
print(df_merged['is_outlier_totalcharges'].value_counts())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def marcar_outliers(df, coluna, plotar=False):
    """
    Identifica outliers com base no m√©todo IQR e marca no DataFrame.

    Par√¢metros:
        df (DataFrame): DataFrame de entrada.
        coluna (str): Nome da coluna num√©rica a ser analisada.
        plotar (bool): Se True, exibe um boxplot comparando outliers.

    Retorno:
        DataFrame com nova coluna 'is_outlier_<coluna>'
    """
    # Calcular Q1, Q3 e IQR
    Q1 = df[coluna].quantile(0.25)
    Q3 = df[coluna].quantile(0.75)
    IQR = Q3 - Q1

    # Limites inferior e superior
    lim_inf = Q1 - 1.5 * IQR
    lim_sup = Q3 + 1.5 * IQR

    # Nome da nova coluna
    nome_col = f'is_outlier_{coluna.lower()}'

    # Marcar outliers
    df[nome_col] = ~df[coluna].between(lim_inf, lim_sup)

    # Exibir boxplot se solicitado
    if plotar:
        plt.figure(figsize=(6,4))
        sns.boxplot(x=nome_col, y=coluna, data=df)
        plt.title(f'Distribui√ß√£o de {coluna} por presen√ßa de Outlier')
        plt.xlabel('√â outlier?')
        plt.ylabel(coluna)
        plt.show()
    
    return df
```

```python
# Para totalcharges com visualiza√ß√£o
df_merged = marcar_outliers(df_merged, 'totalcharges', plotar=True)

# Para monthlycharges sem gr√°fico
df_merged = marcar_outliers(df_merged, 'monthlycharges')
```

üí° Detec√ß√£o de outliers por grupo (por tipo de contrato)
Se quiser marcar os outliers separadamente dentro de cada grupo, aqui vai uma fun√ß√£o adaptada:

```python
def detectar_outliers_por_grupo(df, grupo, coluna):
    """
    Adiciona uma coluna indicando se √© outlier com base no IQR, por grupo.
    """
    nome_col = f'outlier_{coluna.lower()}'
    
    df[nome_col] = False
    
    for grupo_valor, grupo_df in df.groupby(grupo):
        Q1 = grupo_df[coluna].quantile(0.25)
        Q3 = grupo_df[coluna].quantile(0.75)
        IQR = Q3 - Q1
        lim_inf = Q1 - 1.5 * IQR
        lim_sup = Q3 + 1.5 * IQR
        
        idx_outliers = grupo_df[
            (grupo_df[coluna] < lim_inf) | (grupo_df[coluna] > lim_sup)
        ].index
        
        df.loc[idx_outliers, nome_col] = True

    return df
```

Passo a passo do m√©todo IQR:
Calcular o Q1 (1¬∫ quartil) e Q3 (3¬∫ quartil).

Calcular o IQR = Q3 - Q1.

Definir os limites:

Limite inferior = Q1 - 1.5 √ó IQR

Limite superior = Q3 + 1.5 √ó IQR

Os valores fora desses limites s√£o considerados outliers.

```python
import pandas as pd

# Criar c√≥pia do DataFrame para n√£o alterar o original
df_outliers = df_merged[['contract', 'totalcharges']].copy()

# Garantir que 'totalcharges' esteja no formato num√©rico
df_outliers['totalcharges'] = pd.to_numeric(df_outliers['totalcharges'], errors='coerce')

# Remover valores nulos
df_outliers = df_outliers.dropna()

# Lista para armazenar outliers
outliers_detectados = []

# Detectar outliers por tipo de contrato
for contrato in df_outliers['contract'].unique():
    grupo = df_outliers[df_outliers['contract'] == contrato]
    
    Q1 = grupo['totalcharges'].quantile(0.25)
    Q3 = grupo['totalcharges'].quantile(0.75)
    IQR = Q3 - Q1

    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR

    outliers = grupo[(grupo['totalcharges'] < limite_inferior) | (grupo['totalcharges'] > limite_superior)]
    outliers_detectados.append(outliers)

# Concatenar todos os outliers detectados em um √∫nico DataFrame
df_outliers_final = pd.concat(outliers_detectados)

# Exibir os 10 primeiros outliers
print(df_outliers_final.head(10))

# Ou, se quiser salvar em um arquivo CSV para abrir no Excel
df_outliers_final.to_csv("outliers_totalcharges_por_contrato.csv", index=False)
```

Para complementar essa aula, segue o c√≥digo completo para detectar outliers com Z-Score, ideal para distribui√ß√µes aproximadamente normais

```python
import pandas as pd
import numpy as np
from scipy.stats import zscore

# Suponha que df_merged esteja carregado e com colunas num√©ricas convertidas corretamente
colunas_numericas = ['totalcharges', 'monthlycharges', 'tenure']

# Aplica o Z-Score nas colunas selecionadas (ignora nulos)
z_scores = df_merged[colunas_numericas].apply(zscore)

# Converte para valores absolutos
z_abs = np.abs(z_scores)

# Define limite para ser considerado outlier
limite = 3

# Detecta linhas com pelo menos um z-score acima do limite
outliers_zscore = df_merged[(z_abs > limite).any(axis=1)]

# Resultado
print(f'Total de outliers detectados com Z-Score: {len(outliers_zscore)}')

# Exporta para CSV
outliers_zscore.to_csv('outliers_zscore.csv', index=False)
```
