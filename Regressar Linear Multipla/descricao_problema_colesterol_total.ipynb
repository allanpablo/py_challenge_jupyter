# 🩺 Descrição do Problema – Previsão de Colesterol Total

## 🎯 Objetivo
Desenvolver um **modelo de regressão linear múltipla** para prever o valor do **colesterol total** com base em características clínicas de pacientes:
- Idade
- Nível de atividade física
- Peso corporal

## 🏥 Contexto
Este projeto é destinado a um **laboratório de análises clínicas**, com o objetivo de:
- Compreender como variáveis clínicas influenciam nos resultados de colesterol
- Realizar estimativas antecipadas com base em dados de triagem
- Otimizar o atendimento e oferecer diagnósticos preventivos

## 🔧 Etapas do Projeto
1. **Carregamento dos dados**
2. **Análise exploratória (EDA)**
3. **Preparação do dataset (X e y)**
4. **Treinamento do modelo de regressão múltipla**
5. **Validação com métricas: MAE, MSE, R²**
6. **Análise de resíduos e normalidade**
7. **Exportação do modelo**
8. *(Opcional)* Interface para uso prático com dados clínicos

## 📐 Equação da Regressão Múltipla
\[
\text{Colesterol Total} = \beta_0 + \beta_1 \cdot \text{Idade} + \beta_2 \cdot \text{Atividade Física} + \beta_3 \cdot \text{Peso} + \varepsilon
\]
