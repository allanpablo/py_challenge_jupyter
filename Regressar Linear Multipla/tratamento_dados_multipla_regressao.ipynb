# Tratamento de Dados - Regressão Linear Múltipla

## Carregamento do Dataset

```python
import pandas as pd

df = pd.read_csv('1701111759794-attachment.csv')
df.head()
```

## Renomeando colunas para nomes técnicos

```python
df.rename(columns={
    'Id': 'id',
    'Grupo Sanguíneo': 'grupo_sanguineo',
    'Fumante': 'fumante',
    'Nível de Atividade': 'atividade_fisica',
    'Idade': 'idade',
    'Peso': 'peso',
    'Altura': 'altura',
    'Colesterol': 'colesterol_total'
}, inplace=True)
df.head()
```

## Preenchendo valores ausentes e convertendo tipos

```python
df['idade'] = df['idade'].fillna(df['idade'].median()).astype(int)
df['peso'] = df['peso'].fillna(df['peso'].median())
df['altura'] = df['altura'].fillna(df['altura'].median()).astype(int)
df['atividade_fisica'] = df['atividade_fisica'].fillna(df['atividade_fisica'].mode()[0])
df['colesterol_total'] = df['colesterol_total'].fillna(df['colesterol_total'].median())
df.info()
```
