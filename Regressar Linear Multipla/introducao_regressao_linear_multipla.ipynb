# 📘 Introdução à Regressão Linear Múltipla

Neste material, vamos recapitular o conceito de **regressão linear simples** e introduzir a **regressão linear múltipla**, que será aplicada em nosso projeto com múltiplas variáveis independentes para prever uma variável dependente.

## 🔹 Regressão Linear Simples

A regressão linear simples é usada para modelar a relação entre uma variável independente \( x \) e uma variável dependente \( y \), utilizando a equação de uma reta:

\[ y = \beta_0 + \beta_1 x + \varepsilon \]

- \( \beta_0 \): intercepto (onde a reta cruza o eixo Y)
- \( \beta_1 \): coeficiente que representa o impacto de \( x \) sobre \( y \)
- \( \varepsilon \): erro ou resíduo da predição


## 🔹 Regressão Linear Múltipla

A regressão linear múltipla estende o conceito anterior para lidar com **mais de uma variável independente**:

\[ y = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \dots + \beta_n x_n + \varepsilon \]

- Cada \( x_i \) representa uma característica diferente.
- Cada \( \beta_i \) é um coeficiente ajustado pelo modelo.
- O modelo busca o **hiperplano** que melhor se ajusta aos dados, minimizando os erros quadráticos (OLS — Ordinary Least Squares).

## 📐 Visualização

- **Regressão simples**: reta 2D
- **Regressão múltipla**: plano ou hiperplano em 3D ou mais dimensões

Embora não possamos visualizar diretamente mais de 3 dimensões, o conceito é matematicamente o mesmo: encontrar os coeficientes que minimizam a diferença entre os valores reais e os previstos.

## 🎯 Objetivo

- Criar um modelo que estime o valor de \( y \) com base em múltiplas entradas \( x_1, x_2, ..., x_n \)
- Avaliar a performance com métricas como MAE, MSE e R²
- Integrar esse modelo em uma **interface de usuário** para facilitar o uso prático no contexto do projeto
