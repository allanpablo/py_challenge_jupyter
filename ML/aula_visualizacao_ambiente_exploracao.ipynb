## üì¶ 2. Importando bibliotecas


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

## üìÑ 3. Gerando um dataset fict√≠cio
> Como os dados s√£o aleat√≥rios e n√£o controlados, simularemos um conjunto com vari√°veis comuns em an√°lises preditivas.

```python
np.random.seed(42)
```

```python
df = pd.DataFrame({
    "idade": np.random.randint(20, 70, 100),
    "peso": np.random.normal(70, 15, 100),
    "altura": np.random.normal(1.65, 0.1, 100),
    "pressao_sistolica": np.random.randint(100, 160, 100),
    "diabetico": np.random.choice([0, 1], size=100)
})
```

```python
df.head()
```

## üìä 4. An√°lise inicial dos dados

```python
df.describe()
```

```python
df.info()
```

## üìà 5. Visualiza√ß√£o com Seaborn

### Distribui√ß√£o das vari√°veis

```python
sns.histplot(df["idade"], kde=True)
plt.title("Distribui√ß√£o de Idade")
plt.show()
sns.histplot(df["peso"], kde=True)
plt.title("Distribui√ß√£o de Peso")
plt.show()
```

### Rela√ß√£o entre vari√°veis

```python
sns.scatterplot(x="idade", y="pressao_sistolica", hue="diabetico", data=df)
plt.title("Idade vs Press√£o Sist√≥lica (Colorido por Diabetes)")
plt.show()
sns.pairplot(df, hue="diabetico")
```

## üßÆ 6. Correla√ß√£o

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matriz de Correla√ß√£o")
plt.show()
```
