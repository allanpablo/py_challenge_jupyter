## üì¶ Etapa 1: Importa√ß√£o das bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## üìÇ Etapa 2: Carregando os dados

```python
# O arquivo deve estar na mesma pasta do notebook
df = pd.read_csv("exame_diabes.csv")
df.head(10)
```

## ‚ÑπÔ∏è Etapa 3: Estrutura dos dados

```python
df.info()
```

```python
df.describe()
```

## üìä Etapa 4: Distribui√ß√£o de vari√°veis

```python
sns.histplot(df["idade"], kde=True)
plt.title("Distribui√ß√£o de Idade")
plt.show()
sns.histplot(df["peso"], kde=True)
plt.title("Distribui√ß√£o de Peso")
plt.show()
```

```python
sns.histplot(df["altura"], kde=True)
plt.title("Distribui√ß√£o de Altura")
plt.show()
```

```python
sns.histplot(df["resultado"], kde=True)
plt.title("Distribui√ß√£o dos Resultados de Exame")
plt.show()
```

## üîç Etapa 5: Rela√ß√£o entre vari√°veis

```python
sns.scatterplot(x="idade", y="resultado", hue="genero", data=df)
plt.title("Idade vs Resultado de Exame")
plt.show()
sns.scatterplot(x="peso", y="resultado", hue="genero", data=df)
plt.title("Peso vs Resultado de Exame")
plt.show()
```

## üßÆ Etapa 6: Correla√ß√£o

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="viridis", fmt=".2f")
plt.title("Matriz de Correla√ß√£o entre Vari√°veis")
plt.show()
```
