## ðŸŽ¯ Objetivo
Entender de forma visual como funciona a regressÃ£o linear utilizando `sns.regplot`.

## ðŸ“¦ Etapa 1: Importar bibliotecas

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## ðŸ§ª Etapa 2: Criar dataset fictÃ­cio (tempo de casa vs salÃ¡rio)

```python
df = pd.DataFrame({
    "tempo_casa": [1, 2, 3, 4, 5, 6, 7, 8],
    "salario": [2500, 2700, 3000, 3300, 3500, 3700, 4000, 4200]
})
df
```

## ðŸ“‰ Etapa 3: Visualizar regressÃ£o linear

```python
sns.regplot(x="tempo_casa", y="salario", data=df)
plt.title("RegressÃ£o Linear - Tempo de Casa vs SalÃ¡rio")
plt.xlabel("Tempo de Casa (anos)")
plt.ylabel("SalÃ¡rio (R$)")
plt.show()
```

## ðŸ“š Etapa 4: IntuiÃ§Ã£o

A linha de regressÃ£o Ã© uma estimativa da relaÃ§Ã£o entre as duas variÃ¡veis:

```python
- Quanto maior o tempo de casa, maior tende a ser o salÃ¡rio
- A reta foi ajustada minimizando a distÃ¢ncia dos pontos reais atÃ© a linha (erro)
```

## âœ… ConclusÃ£o

- `sns.regplot()` mostra a relaÃ§Ã£o entre variÃ¡veis e a linha de melhor ajuste
- Essa abordagem Ã© fundamental para prever valores com base em tendÃªncias
- A regressÃ£o linear Ã© a base para diversos modelos preditivos
